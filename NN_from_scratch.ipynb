```python
def generate_w(num_layer, input_size,num_nodes,output_size):
  input_layer = [[0 for _ in range(input_size)] for _ in range(num_nodes)]
  input_layer2 = [[0 for _ in range(num_nodes)] for _ in range(num_nodes)]
  w=[]
  w.append(input_layer)
  for i in range (num_layer-3):
    w.append(input_layer2)

  outputlayer = [[0 for _ in range(num_nodes)] for _ in range(output_size)]
  w.append(outputlayer)
  return w
```

```python
def generate_Nodes(hiden_layer_size,output_layer_size,input_sample_size,num_of_layers):

  Nodes=[]
  input_layer = [[0 for _ in range(3)] for _ in range(input_sample_size)]
  Nodes.append(input_layer)
  for i in range (num_of_layers-2):
    hiden_layer=[[0 for _ in range(3)] for _ in range(hiden_layer_size)]
    Nodes.append(hiden_layer)


  output_layer=[[0 for _ in range(3)] for _ in range(output_layer_size)]
  Nodes.append(output_layer)

  return Nodes
```

```python
def sigmoid(input):
  return 1 / (1 + np.exp(-input))
```

```python
def forward(W,Nodes):

  rows = len(Nodes)       #
  cols = len(Nodes[0])    #3

  for row in range (len(Nodes)-1) :

    for nod in range(len(Nodes[row+1])):

      for i in range(len(Nodes[row])):

        Nodes[row+1][nod][0]=Nodes[row+1][nod][0]+Nodes[row][i][1]*W[row][nod][i]
      Nodes[row+1][nod][1]=sigmoid(Nodes[row+1][nod][0])

  return Nodes
```

```python

def backword(nodes,w,target,eta):
  rows = len(nodes)
  for row in range(rows-1, -1, -1):
    for nod in range(len(nodes[row])):
      if row ==rows-1:
          segma=nodes[row][nod][1]*(1-nodes[row][nod][1])*(target[nod]-nodes[row][nod][1])
          nodes[row][nod][2]=segma
          if row!=0:
            for i in range(len(nodes[row-1])):
              delta_w=eta*nodes[row-1][i][1]*segma
              w[row-1][nod][i]=w[row-1][nod][i]+delta_w

      else:
        for k in range(len(nodes[row+1])):
          sum=nodes[row+1][k][2]*w[row][k][nod]
        segma=nodes[row][nod][1]*(1-nodes[row][nod][1])*sum
        nodes[row][nod][2]=segma
        if row!=0:
          for j in range(len(nodes[row-1])):
            delta_w=eta*nodes[row-1][j][1]*segma
            w[row-1][nod][j]=w[row-1][nod][j]+delta_w


  return(w)
```
